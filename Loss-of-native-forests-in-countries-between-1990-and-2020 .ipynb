```python
# Apresentando somente os dados do ano de 2018 e que sejam maiores ou iguais a 50.00%
# Only showing data for the year 2018 and that are greater than or equal to 50.00% 
data1 = data[(data.index == '2018')][(data >= 50.00)].dropna(axis=1)
data1
```

```python
# Reordenando as linhas e colunas e colocando em ordem descrescente
# Sorting the rows and columns and placing them in descending order 
data1 = data1.pivot_table(columns=['year']).sort_values(by='2018', ascending=False)
data1.head()
```

```python
# Adicionando uma nova coluna (position) com valores de posição 
# Adding a new column (position) with position values 
position = np.arange(len(data1))+1
data1['position'] = position
data1.head()
```

```python
# Apresentando os dados somente do Brasil e reordenando-os do mais antigo ao mais recente
# Displaying data only from Brazil and rearranging from oldest to newest 
brasil = dat.loc['Brazil'].sort_values(by='year')
brasil.head()
```

```python
# adicionando uma coluna chamada 'variation' com a vatiação percentual de ano em ano, com a função 'pct_change' 
# adding a column called 'variation' with the percentage variation from year to year, with the function 'pct_change' 
variation = brasil.pct_change()*100
brasil['variation'] = variation
brasil.head()
```

```python
# calculando a média de perda de florestas nativas no Brasil entre 1990 e 2020
# calculating the average loss of native forests in Brazil between 1990 and 2020
bra = brasil['variation'].loc['1990':'2020'].mean()
#bra
print(f'{bra:.2f}%')
```

```python
# Calculando percentualmente a perda florestal do Brasil de 1990 a 2020
# Calculating the percentage of forest loss in Brazil from 1990 to 2020
value2020 = brasil[brasil.index=='2020']['AG.LND.FRST.ZS'].iloc[0] # retorna o valor percentual de floresta nativa no terrirório brasileiro em 2020
value1990 = brasil[brasil.index=='1990']['AG.LND.FRST.ZS'].iloc[0] # retorna o valor percentual de floresta nativa no terrirório brasileiro em 1990
per = ((value2020-value1990)/value1990)*100 # calcula quanto de floresta nativa foi perdido entre 1990 e 2020
print(f'{per:.2f}%')
```

```python
# Modificando as linhas e colunas, colocando os anos como índices e os países como colunas
# Modifying rows and columns, putting years as indices and countries as columns 
data2 = dat.pivot_table(values='AG.LND.FRST.ZS', index=['country'], columns=['year'])
data2
```

```python
data2.reset_index(inplace=True)
data2
```

```python
data2[['country', '2018']]
```

```python
# Calculando percentualmente a perda/ganho de florestal dos países de 1990 a 2020
# Percentage calculation of forest loss/gain for countries from 1990 to 2020
world2020 = data2[['country','2020']] # retorna o valor percentual de floresta nativa nos países em 2020
world1990 = data2[['country','1990']] # retorna o valor percentual de floresta nativa nos países em 1990
per = ((world2020['2020']-world1990['1990'])/world1990['1990'])*100 # calcula percentualmente quanto de floresta nativa foi perdido ou ganhado entre 1990 e 2020
data2['percentual'] = per # acrescenta uma coluna com o percentual de perda/ganho de floresta nativa nos países

# criando um dataframe novo só com os países e o pencentual de perda/ganho entre 1990 e 2020 em ordem decrescente de peda
dataper1 = data2[['country', 'percentual']].sort_values(by='percentual', ascending=True)

# criando um dataframe novo só com os países e o pencentual de perda/ganho entre 1990 e 2020 em ordem decrescente de ganho
dataper2 = data2[['country', 'percentual']].sort_values(by='percentual', ascending=False)

dataper2
```

```python
# Adicionando uma nova coluna (position) com valores de posição no dataper1 
# Adding a new column (position) with position values in dataper1
position = np.arange(len(dataper1))+1
dataper1['position'] = position
dataper1.set_index('position', inplace=True) # transformando a nova coluna no índice
dataper1.iloc[:10]
```

```python
# Adicionando uma nova coluna (position) com valores de posição no dataper2
# Adding a new column (position) with position values in dataper2
position = np.arange(len(dataper2))+1
dataper2['position'] = position
dataper2.set_index('position', inplace=True) # transformando a nova coluna no índice
dataper2.iloc[:10]
```
